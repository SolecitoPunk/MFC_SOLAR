# Solución Analítica de la Ecuación de Schrödinger  
## Potencial de Doble Pozo: $V(x) = 0.1(x^2 - 10)^2$

---

## Planteamiento del problema

La ecuación de Schrödinger independiente del tiempo es:

$$-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2} + V(x)\psi = E\psi$$

En **unidades naturales** ($\hbar = m = 1$), con el factor $\frac{1}{2}$ del código Fortran:

$$\boxed{-\frac{1}{2}\frac{d^2\psi}{dx^2} + V(x)\psi = E\psi}$$

con el potencial:

$$V(x) = 0.1(x^2 - 10)^2$$

---

## Paso 1: Análisis del potencial

Expandiendo el potencial:

$$V(x) = 0.1(x^4 - 20x^2 + 100) = 0.1x^4 - 2x^2 + 10$$

### Puntos críticos

Derivando e igualando a cero:

$$V'(x) = 0.4x^3 - 4x = 0 \implies x\,(0.4x^2 - 4) = 0$$

Las soluciones son:

$$x = 0, \qquad x = \pm\sqrt{10} \approx \pm 3.162$$

Evaluando $V''(x) = 1.2x^2 - 4$:

| Punto | $V''$ | Tipo | $V(x)$ |
|-------|-------|------|--------|
| $x = 0$ | $-4 < 0$ | Máximo (barrera) | $V(0) = 10$ |
| $x = \pm\sqrt{10}$ | $+8 > 0$ | Mínimo (pozo) | $V(\pm\sqrt{10}) = 0$ |

El potencial tiene **dos pozos en $x = \pm\sqrt{10}$** separados por una **barrera de altura $V_0 = 10$** en $x = 0$.

---

## Paso 2: Aproximación armónica cerca de los mínimos

Como no existe solución analítica exacta para este potencial, se aproxima localmente mediante una **expansión en serie de Taylor** alrededor de $x_0 = +\sqrt{10}$.

Definiendo el desplazamiento $u = x - \sqrt{10}$:

$$V(x_0 + u) \approx V(x_0) + \frac{1}{2}V''(x_0)\,u^2 + \frac{1}{6}V'''(x_0)\,u^3 + \frac{1}{24}V^{(4)}(x_0)\,u^4 + \cdots$$

Calculando las derivadas en $x_0 = \sqrt{10}$:

$$V(x_0) = 0$$

$$V''(x_0) = 1.2(10) - 4 = 8$$

$$V'''(x_0) = 2.4\sqrt{10}$$

$$V^{(4)}(x_0) = 2.4$$

A **orden cuadrático** (aproximación armónica):

$$\boxed{V(x) \approx 4u^2}$$

---

## Paso 3: Oscilador armónico local

Con la aproximación armónica, la ecuación de Schrödinger se convierte en:

$$-\frac{1}{2}\frac{d^2\psi}{du^2} + 4u^2\psi = E\psi$$

Comparando con el oscilador armónico estándar:

$$-\frac{1}{2}\frac{d^2\psi}{du^2} + \frac{1}{2}\omega^2 u^2\psi = E\psi$$

se identifica:

$$\frac{1}{2}\omega^2 = 4 \implies \boxed{\omega = 2\sqrt{2} \approx 2.828}$$

### Energías del oscilador armónico local

$$E_n^{(\text{loc})} = \omega\left(n + \frac{1}{2}\right) = 2\sqrt{2}\left(n + \frac{1}{2}\right)$$

| $n$ | $E_n^{(\text{loc})}$ | $E_n^{(\text{numérico})}$ |
|-----|---------------------|--------------------------|
| 0 | 1.414 | 0.973 |
| 1 | 4.243 | 0.973 |
| 2 | 7.071 | 2.796 |

La diferencia entre los valores locales y los numéricos se debe a los **términos ánarmónicos** (cúbico y cuártico) despreciados en la aproximación.

---

## Paso 4: Funciones de onda del oscilador armónico

Para el pozo en $x_0 = +\sqrt{10}$, las soluciones exactas son:

$$\boxed{\psi_n^{(+)}(u) = N_n\, H_n\!\left(\sqrt{\omega}\, u\right) e^{-\frac{\omega u^2}{2}}}$$

donde $H_n$ son los **polinomios de Hermite** y $N_n$ es la constante de normalización:

$$N_n = \left(\frac{\omega}{\pi}\right)^{1/4} \frac{1}{\sqrt{2^n\, n!}}$$

Los primeros polinomios de Hermite son:

$$H_0(y) = 1$$

$$H_1(y) = 2y$$

$$H_2(y) = 4y^2 - 2$$

$$H_3(y) = 8y^3 - 12y$$

Así, las primeras funciones de onda locales son:

$$\psi_0^{(+)}(u) = \left(\frac{\omega}{\pi}\right)^{1/4} e^{-\frac{\omega u^2}{2}}$$

$$\psi_1^{(+)}(u) = \left(\frac{\omega}{\pi}\right)^{1/4} \sqrt{2\omega}\; u\; e^{-\frac{\omega u^2}{2}}$$

---

## Paso 5: Desdoblamiento de niveles y efecto túnel

Debido a la **simetría** del potencial $V(x) = V(-x)$, el sistema tiene soluciones simétricas (pares) y antisimétricas (impares). Las soluciones globales se construyen como combinaciones lineales de las soluciones locales de cada pozo:

$$\psi^{(+)} = \frac{1}{\sqrt{2}}\left[\psi_n^{(+)}(x) + \psi_n^{(-)}(x)\right] \qquad \text{(simétrica, par)}$$

$$\psi^{(-)} = \frac{1}{\sqrt{2}}\left[\psi_n^{(+)}(x) - \psi_n^{(-)}(x)\right] \qquad \text{(antisimétrica, impar)}$$

Esto produce un **desdoblamiento** de cada nivel del oscilador local en **dos niveles** del sistema completo. El desdoblamiento en energía es:

$$\Delta E_n = E_n^{(-)} - E_n^{(+)} \propto e^{-S}$$

donde $S$ es la **acción bajo la barrera** (resultado WKB), calculada en el paso siguiente.

---

## Paso 6: Aproximación WKB y cálculo de la acción

La aproximación **WKB** (Wentzel–Kramers–Brillouin) permite calcular la probabilidad de efecto túnel a través de la barrera. La acción bajo la barrera es:

$$S = \int_{-\sqrt{10}}^{+\sqrt{10}} \sqrt{2\left[V(x) - E_n\right]}\; dx$$

Para los estados más bajos ($E_n \approx 0$):

$$S \approx \int_{-\sqrt{10}}^{+\sqrt{10}} \sqrt{2 \cdot 0.1\,(x^2 - 10)^2}\; dx = \sqrt{0.2}\int_{-\sqrt{10}}^{+\sqrt{10}} (10 - x^2)\; dx$$

Calculando la integral:

$$\int_{-\sqrt{10}}^{+\sqrt{10}} (10 - x^2)\; dx = \left[10x - \frac{x^3}{3}\right]_{-\sqrt{10}}^{+\sqrt{10}} = 2\left(10\sqrt{10} - \frac{10\sqrt{10}}{3}\right) = \frac{40\sqrt{10}}{3}$$

Por lo tanto:

$$\boxed{S \approx \sqrt{0.2} \cdot \frac{40\sqrt{10}}{3} \approx 18.86}$$

El desdoblamiento de niveles es entonces:

$$\Delta E \sim e^{-S} = e^{-18.86} \approx 6 \times 10^{-9}$$

Este valor extremadamente pequeño explica por qué los pares de niveles aparecen **casi degenerados** en los resultados numéricos:

$$E_0 \approx E_1 \approx 0.973, \qquad E_2 \approx E_3 \approx 2.796$$

---

## Resumen

| Paso | Método | Resultado principal |
|------|--------|-------------------|
| 1 | Análisis del potencial | Dos mínimos en $x = \pm\sqrt{10}$, barrera $V_0 = 10$ en $x=0$ |
| 2 | Serie de Taylor | Aproximación local: $V \approx 4u^2$ |
| 3 | Oscilador armónico | Frecuencia local: $\omega = 2\sqrt{2}$ |
| 4 | Polinomios de Hermite | $\psi_n \sim H_n(\sqrt{\omega}\,u)\,e^{-\omega u^2/2}$ |
| 5 | Simetría del potencial | Desdoblamiento en pares simétrico/antisimétrico |
| 6 | WKB | $S \approx 18.86 \Rightarrow \Delta E \sim e^{-18.86} \approx 0$ |

### Conclusión física

Este potencial **no tiene solución analítica cerrada exacta**. La combinación de aproximación armónica + WKB + simetría captura la física esencial:

- Los niveles se agrupan en **pares casi degenerados** debido al efecto túnel despreciable.
- Las funciones de onda están **localizadas en cada pozo** (estados ligados) o **delocalizadas simétricamente** entre ambos pozos.
- La barrera central actúa como separador clásico, pero la mecánica cuántica permite la **tunelización** entre pozos.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from scipy.linalg import eigh_tridiagonal

# ── Parámetros ──────────────────────────────────────────────
xmin, xmax = -10, 10
N = 1000
x = np.linspace(xmin, xmax, N)
h = x[1] - x[0]
V = 0.1 * (x**2 - 10)**2

# ── Método de diferencias finitas (igual que Fortran) ───────
def solver_fortran_style(x, V, h):
    diag = 2/(2*h**2) + V          # factor 1/2 de hbar²/2m
    offdiag = -1/(2*h**2) * np.ones(N-1)
    vals, vecs = eigh_tridiagonal(diag, offdiag, eigvals_only=False)
    # Normalizar
    for i in range(vecs.shape[1]):
        norm = np.trapz(vecs[:,i]**2, x)
        vecs[:,i] /= np.sqrt(norm)
    return vals, vecs

# ── Método de Numerov ────────────────────────────────────────
def numerov_shoot(E, x, V, h):
    """Integra desde la izquierda con energía E y devuelve psi"""
    f = 2*(V - E)
    psi = np.zeros(len(x))
    psi[0] = 0.0
    psi[1] = 1e-6   # valor pequeño para arrancar
    for i in range(1, len(x)-1):
        num = 2*psi[i]*(1 - 5*h**2/12 * f[i]) - psi[i-1]*(1 + h**2/12 * f[i-1])
        den = 1 + h**2/12 * f[i+1]
        psi[i+1] = num / den
    return psi

def encontrar_energia_numerov(E_min, E_max, x, V, h, n_estados=6, tol=1e-8):
    """Busca autoenergías por shooting method + bisección"""
    from scipy.optimize import brentq

    def boundary(E):
        psi = numerov_shoot(E, x, V, h)
        return psi[-1]  # condición de frontera: psi(xmax) = 0

    # Barrido grueso para encontrar cambios de signo
    E_scan = np.linspace(E_min, E_max, 5000)
    vals_boundary = [boundary(E) for E in E_scan]

    energias = []
    psis = []
    for i in range(len(E_scan)-1):
        if vals_boundary[i] * vals_boundary[i+1] < 0:
            # Cambio de signo → hay un autovalor aquí
            E_found = brentq(boundary, E_scan[i], E_scan[i+1], xtol=tol)
            psi = numerov_shoot(E_found, x, V, h)
            # Normalizar
            norm = np.trapz(psi**2, x)
            psi /= np.sqrt(abs(norm))
            # Signo consistente
            if psi[N//2] < 0:
                psi = -psi
            energias.append(E_found)
            psis.append(psi)
            if len(energias) == n_estados:
                break

    return np.array(energias), np.array(psis)

print("Calculando diferencias finitas (estilo Fortran)...")
E_fortran, psi_fortran = solver_fortran_style(x, V, h)

print("Calculando Numerov...")
E_numerov, psi_numerov = encontrar_energia_numerov(0.1, 6.0, x, V, h, n_estados=6)

print("\n{'Estado':<8} {'E_Fortran':>12} {'E_Numerov':>12} {'Diferencia':>12}")
print("-" * 48)
for i in range(6):
    dif = abs(E_fortran[i] - E_numerov[i])
    print(f"  n={i}    {E_fortran[i]:>12.6f} {E_numerov[i]:>12.6f} {dif:>12.2e}")

# ── Leer datos del Fortran real ──────────────────────────────
with open("data_oscilador") as f:
    primera_linea = f.readline()
    valores = primera_linea.strip("# \n").split()
    E_fortran_archivo = np.array([float(v) for v in valores[1:]])

data = np.loadtxt("data_oscilador", comments="#")
x_fort = data[:, 0]
psi2_fort = data[:, 1:7]

# ── Gráfico interactivo ──────────────────────────────────────
colores = ["tab:blue", "tab:orange", "tab:green",
           "tab:red",  "tab:purple", "tab:brown"]

def graficar(n):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # ── Panel izquierdo: comparación de densidades ───────────
    ax = axes[0]
    ax.plot(x, V, 'k-', linewidth=1.5, alpha=0.4, label="V(x)")

    E_f = E_fortran[n]
    E_num = E_numerov[n]
    escala = 0.4

    ax.plot(x, psi_fortran[:, n]**2 * escala + E_f,
            color=colores[n], linewidth=2, label=f"Dif. Finitas  E={E_f:.5f}")
    ax.plot(x, psi_numerov[n]**2 * escala + E_num,
            color=colores[n], linewidth=1.5, linestyle="--",
            label=f"Numerov  E={E_num:.5f}")

    ax.hlines(E_f,   xmin, xmax, colors="gray",  linestyles="dashed", linewidth=0.8)
    ax.hlines(E_num, xmin, xmax, colors="orange", linestyles="dotted", linewidth=0.8)

    ax.fill_between(x, E_f, psi_fortran[:, n]**2 * escala + E_f,
                    alpha=0.15, color=colores[n])

    ax.set_xlim(xmin, xmax)
    ax.set_ylim(-0.5, 6)
    ax.set_xlabel("x")
    ax.set_ylabel("Energía")
    ax.set_title(f"n={n} — Comparación de métodos")
    ax.legend(fontsize=8)
    ax.grid(alpha=0.3)

    # ── Panel derecho: diferencia entre métodos ───────────────
    ax2 = axes[1]
    diff = psi_fortran[:, n]**2 - psi_numerov[n]**2
    ax2.plot(x, diff, color=colores[n], linewidth=1.2)
    ax2.axhline(0, color='k', linewidth=0.8, linestyle='--')
    ax2.set_xlabel("x")
    ax2.set_ylabel(r"$|\psi_{DF}|^2 - |\psi_{Num}|^2$")
    ax2.set_title(f"n={n} — Diferencia entre métodos")
    ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

out = widgets.interactive(
    graficar,
    n=widgets.IntSlider(value=0, min=0, max=5, step=1,
                        description="Estado n:",
                        continuous_update=False,
                        layout=widgets.Layout(width="500px"))
)

display(out)
```
